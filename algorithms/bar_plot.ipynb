```python
def line_plot(x,y,y_label,name):
  import seaborn as sns
  import matplotlib.pyplot as plt
  front_size = 15

  sns.set_style("whitegrid")

  # Create the line plot using Seaborn
  fig, ax = plt.subplots()
  sns.lineplot(x=x, y=y, marker="o", markersize=10, color="teal", linewidth=2)

  # Add labels and title
  plt.ylabel(f"{y_label}",fontsize = front_size)
  # plt.title("Fancy Line Plot")

  # Customize the axes limits, if needed
  # plt.xlim(0, 6)
  # plt.ylim(0, 12)

  # Show the plot
  plt.tight_layout()
  plt.savefig(name)
  plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a Pandas DataFrame called 'df' with multiple columns
# Let's say you want to plot columns 'column1', 'column2', and 'column3'

# Plotting the bar plot
Exkmc = [0.7143, 0.3077, 0, 0.7059, 0.3333, 0.6667, 0, 0.6154, 0.6667, 0.4286, 0.4439]
icot = [0.3636, 0.5000, 0.6316, 0.8333, 0.5000, 0, 0.4000, 0.6667, 0.6667, 0.4444, 0.5006]
cart = [0.4444, 0.3333, 0, 0.6154, 0.4000, 0.3333, 0, 0.2500, 0.4000, 0.5000, 0.3276]
index =['Airpush','BankBot','Dowgin','DroidKungFu', 'FakeInst', 'Fusob', 'Jisut', 'Kuguo','Mecor','Youmi','Average']
d = {'Exkmc':Exkmc, 'ICOT':icot,'CART': cart,'family':index}

df = pd.DataFrame(d)


# sns.barplot(x= 'family', hue= '', data=df)
# plt.show()
fig, ax = plt.subplots(figsize=(30,10))
# x = np.arrange(len(top5_alcohol))
width = 0.15
x = np.arange(len(df.index))
plt.bar(x-0.2, df['Exkmc'],
        width, color='c', label='Exkmc')
plt.bar(x, df['ICOT'],
        width, color='orange', label='ICOT')
plt.bar(x+0.2, df['CART'],
        width, color='gray', label='CART')
# plt.title('TOP5 countries by pure alcohol consumption', fontsize=25)
font_size = 25

plt.xlabel(None)
plt.ylabel('F1score',fontsize=font_size)
plt.xticks(df.index, df['family'], fontsize=font_size)
# plt.ylabel('Servings per person', fontsize=20)
plt.yticks(fontsize=font_size)
sns.despine(bottom=True)
ax.grid(True,axis='y', zorder=0)
ax.tick_params(bottom=False, left=True)
plt.legend(frameon=False, fontsize=font_size)
plt.savefig('fcore_algs')
plt.show()
```
